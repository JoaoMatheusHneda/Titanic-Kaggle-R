```python
library(caret)
trControl <- trainControl(method = "repeatedcv",
                            number = 10,
                            ## 10-fold CV
                            repeats = 3,
                            ## repeated three times
                            classProbs = TRUE,
                            savePredictions = TRUE,
                            summaryFunction = defaultSummary,
                            returnData = TRUE)
```

```python
caret_model <- train(Survived ~ .,
                       data = train_tidy_ml,
                       method = "glmStepAIC",
                       trControl = trControl,
                       family = binomial(link="logit"),
                       metric = "Accuracy",
                       direction = "both")
```
